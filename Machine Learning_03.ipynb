```python
import pandas as pd
df=pd.read_csv(r"D:\PYTHON CODES\Excel_Pandas\Machine_Learning03.csv")
df
```

```python
dumies=pd.get_dummies(df.town)
dumies
```

```python
merged=pd.concat([df,dumies],axis='columns')
merged
```

```python
final=merged.drop(['town','delhi'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()

```

```python
X=final.drop('price',axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[2800,1,0]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```
