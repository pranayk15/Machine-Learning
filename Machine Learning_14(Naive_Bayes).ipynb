```python
import pandas as pd
df=pd.read_csv(r"D:\\PYTHON CODES\\Excel_Pandas\\Machine_Learning06.csv")
df
```

```python
df.drop(['passenger id','name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df
```

```python
df.age
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

```python
dummies=pd.get_dummies(inputs.sex)
dummies
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs
```

```python
inputs.drop('sex',axis='columns',inplace=True)
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.age
```

```python
inputs.age=inputs.age.fillna(inputs.age.mean())
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.9)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
